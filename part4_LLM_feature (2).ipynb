```python
def pii_guardrail(text):

    blocked_words = [
        "phone",
        "mobile",
        "email",
        "aadhaar",
        "pan",
        "address"
    ]

    lower_text = text.lower()

    for word in blocked_words:
        if word in lower_text:
            raise ValueError(f"PII Detected: {word}")

    return text
```

```python
from jsonschema import validate

schema = {
    "type": "object",
    "properties": {
        "prediction_label": {"type": "string"},
        "confidence_level": {"type": "string"},
        "top_reason": {"type": "string"},
        "second_reason": {"type": "string"},
        "next_step": {"type": "string"}
    },
    "required": [
        "prediction_label",
        "confidence_level",
        "top_reason",
        "second_reason",
        "next_step"
    ]
}
```

```python
import requests
import json

API_KEY = "Your_api_key"

url = "https://openrouter.ai/api/v1/chat/completions"

def call_llm(system_prompt, user_prompt, temperature=0):

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "openai/gpt-4o-mini",
        "messages": [
            {"role":"system","content":system_prompt},
            {"role":"user","content":user_prompt}
        ],
        "temperature": temperature
    }

    response = requests.post(url, headers=headers, json=payload)

    response.raise_for_status()

    return response.json()["choices"][0]["message"]["content"]
```

```python
reply = call_llm(
    "Reply using exactly one word.",
    "Hello"
)

print(reply)
```

```python
import json

results = []

system_prompt = """
You are an AI assistant explaining machine learning predictions.

Return ONLY valid JSON in the following format:

{
  "prediction_label": "...",
  "confidence_level": "...",
  "top_reason": "...",
  "second_reason": "...",
  "next_step": "..."
}
"""

for record in records:

    user_prompt = f"""
Features:
{json.dumps(record['features'], indent=2)}

Predicted Class: {record['prediction']}

Predicted Probability: {record['probability']}
"""

    # Apply PII guardrail
    safe_prompt = pii_guardrail(user_prompt)

    # Call LLM
    response = call_llm(system_prompt, safe_prompt, temperature=0)

    response_json = json.loads(response)

    # Validate schema
    validate(instance=response_json, schema=schema)

    results.append(response_json)

print(json.dumps(results, indent=2))
```

```python
comparison = []

record = records[0]

prompt = f"""
Features:
{json.dumps(record['features'], indent=2)}

Prediction:
{record['prediction']}

Probability:
{record['probability']}
"""

temp0 = call_llm(system_prompt, prompt, temperature=0)
temp07 = call_llm(system_prompt, prompt, temperature=0.7)

comparison.append({
    "temperature_0": json.loads(temp0),
    "temperature_0_7": json.loads(temp07)
})

print(json.dumps(comparison, indent=2))
```
